```python
ca_babies = states_babies.loc[states_babies['State']== "CA",:]
```

```python
ca_babies
```

```python
def popular(s):
    """Function receives Pandas series s, containing baby names from highest count to lowest
    count. Returns the most popular name in s."""    
    return s.iloc[0]
```

```python
ca_grouped = ca_babies.sort_values("Count", ascending =False).groupby(["Year","Gender"]).agg(popular)
```
