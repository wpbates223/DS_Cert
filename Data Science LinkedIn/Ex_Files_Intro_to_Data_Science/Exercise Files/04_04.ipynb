```python
# import pandas library 

import pandas as pd
```

```python
# load the data from the csv file into a dataframe

states_babies = pd.read_csv("state_baby_names.csv")
```

```python
# display the dataframe

states_babies.head()
```

```python
# for all the rows where the state is CA, access all the columns
# save the resulting dataframe in a new variable

ca_babies = states_babies.loc[states_babies["State"] == "CA", :]
```

```python
# display the new dataframe

ca_babies
```

```python
# define a function named popular that helps you find the most popular baby name

def popular(s):
    """Receives s, a Pandas Series, containing baby names in order of highest count to lowest count.
    Returns the most popular name in s."""
    
    return s.iloc[0]
```

```python
# sort the baby names in order of highest count to lowest count 
# group the sorted dataframe by the "Year" and "Gender" columns 
# aggregate the groups of data by applying the popular function to every column in each group
# save the grouped dataframe in a new variable

ca_grouped = ca_babies.sort_values("Count", ascending = False).groupby(["Year", "Gender"]).agg(popular)
```

```python
# display the grouped dataframe

ca_grouped
```
